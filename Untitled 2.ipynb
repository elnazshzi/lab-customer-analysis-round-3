```python
# 7.Put all the previously mentioned data transformations into a function.

def process_marketing_data(file_path):
    
    df = pd.read_csv("/Users/elnazshahzeidi/Downloads/marketing_customer_analysis.csv")
    
    
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    
   
    numerical_columns = df.select_dtypes(include=['number']).columns.tolist()
    categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
    
 
    df = df.fillna({
        col: 0 if col in numerical_columns else 'Unknown' for col in df.columns
    })
    
  
    df['effective_to_date'] = pd.to_datetime(df['effective_to_date'])
    
 
    df['month'] = df['effective_to_date'].dt.month
    
   
    first_quarter_data = df[df['month'].isin([1, 2, 3])]
    
    return first_quarter_data


process_marketing_data("/Users/elnazshahzeidi/Downloads/marketing_customer_analysis.csv")
```
